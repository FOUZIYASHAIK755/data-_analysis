```python
import matplotlib.pyplot as plt
```

```python
#line plot in bivariate analysis
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y);
plt.title("simple line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show();

#changing color
plt.plot(x,y,color='red')

#changing line style
plt.plot(x,y,linestyle='dashed')

#adjusting line width
plt.plot(x,y,linewidth=3)

#ADDING MARKERS
plt.plot(x,y,marker='o')

plt.plot(x,y,color='blue',linestyle='dashed',linewidth=2,)
```

```python
x=['jan','feb','mar','apr','may','jun']
y=[500,800,600,200,400,100]
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
plt.xlabel('months(jan-jun)')
plt.ylabel('savings amount')
plt.show()
```

```python
cat=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(cat,values,color='skyblue')
plt.title('bar chart')
plt.xlabel('categories')
plt.ylabel('values')
plt.show()
```

```python
plt.bar(cat,values,color=['red','blue','green','pink','yellow'],edgecolor='black')
plt.show()
```

```python
age=[20,22,15,14,18,12,19,20,16,29,32,25,26,35,40,45,25,35,36,27,37,40,30,20,10,65,70,55,58,62]
b=[20,30,40,50,60,70]
plt.hist(age,bins=b,color='skyblue',edgecolor='black')
plt.title('age distribution of customers')
plt.xlabel('age ranges')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

```python
cat=['A','B','C','D','E']
values=[10,15,7,12,9]
color=['pink','blue','lightblue','green','purple']
plt.pie(values,labels=cat,colors=color,autopct='%1.1f%%')
plt.show()
```

```python
x=[5,10,2,15,20]
y=[10,20,30,40,50]
plt.scatter(x,y,color='blue',s=100,edgecolor='black')
plt.title("scatter plot")
plt.xlabel('x values')
plt.ylabel('y label')
plt.show()
```
