```python
train=pd.read_csv('train.csv')
train
```

```python
sns.barplot(data=train,y='Survived',x='Pclass',palette='magma',errorbar=None)
plt.title('survival rate by passenger class')
plt.xlabel('passenger class')
plt.ylabel('survival rate')
plt.show()
```

```python
sns.scatterplot(data=train,x='Age',y='Fare',hue='Age',palette='magma')
plt.show()
```

```python
subset=train[['Age','Fare','Pclass','Survived']]
sns.pairplot(subset,hue='Survived',palette='viridis',diag_kind='kde')
plt.suptitle('pairwise relationships',y=1.02)
plt.show()
```

```python
train.info()
```

```python
train_data=train.apply(pd.to_numeric,errors='coerce')
train_data.info()
train_num=train_data.corr()
plt.figure(figsize=(12,6))
sns.heatmap(train_num,annot=True,cmap='coolwarm',linewidth=0.5)
plt.title('heatmap')
plt.show()
```
